```python
from PyPDF2 import PdfReader

def ResumeReader(pdf):
  reader = PdfReader(pdf)
  number_of_pages = len(reader.pages)
  text = ""
  for i in range(number_of_pages):
      page = reader.pages[i]
      text += page.extract_text()
  return text
```

## PreProcessing

```python
import pandas as pd

def preprocess_scraped_data(jobs):
  cols_to_remove = ['site', 'job_url', 'job_url_direct', 'date_posted', 'emails', 'company_url', 'company_logo', 'company_url_direct']

  new_jobs = jobs.drop(columns=cols_to_remove, errors='ignore')

  new_jobs['combined_text'] = new_jobs.drop(columns=['id']).astype(str).apply(' '.join, axis=1)

  new_jobs['preprocessed_text'] = new_jobs['combined_text'].apply(preprocess_resume)

  new_jobs = new_jobs.drop_duplicates(subset=['combined_text'])

  return new_jobs
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string
```

```python
def preprocess_resume(text):
  tokens=word_tokenize(text.lower())
  tokens=[word for word in tokens if word not in stop_words and word not in punctuation and word.isalnum()]
  tokens=[lemmatizer.lemmatize(word) for word in tokens]
  return ' '.join(tokens)
```

## KNN Model

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import display
import csv
```

```python
def predict(new_jobs, new_text):
    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(new_jobs['preprocessed_text'])

    def recommend_jobs(new_text, top_n=5):
        candidate_text = new_text
        candidate_vector = vectorizer.transform([candidate_text.lower()])

        similarities = cosine_similarity(candidate_vector, tfidf_matrix)
        sorted_indices = similarities.argsort()[0][::-1]

        top_jobs = new_jobs.iloc[sorted_indices[:top_n]]
        return top_jobs

    recommended_jobs = recommend_jobs(new_text, top_n = 10)
    return recommended_jobs
```

```python
import re
from collections import defaultdict


def extract_skills(text, skills):
    """
    Extract matching skills from the given text based on the predefined skill set.
    Args:
        text (str): The input text to search for skills.
        skills (list): The predefined list of skills.
    Returns:
        list: A list of matching skills found in the text.
    """
    matched_skills = []
    text_lower = text.lower()

    for skill in skills:
        # Check if the skill exists in the text
        if skill.lower() in text_lower:
            matched_skills.append(skill)

    return matched_skills

def match_skills(job_data, resume_text, skill_set):
    """
    Match skills between the job data and the user's resume.
    Args:
        job_data (DataFrame): DataFrame containing job descriptions.
        resume_text (str): The preprocessed text of the user's resume.
        skill_set (list): The predefined skill set.
    Returns:
        dict: Dictionary where each job ID maps to two skill lists:
              - Skills the user has that the job demands.
              - Skills the user doesn't have that the job demands.
    """
    result = {}
    user_skills = extract_skills(resume_text, skill_set)

    for _, row in job_data.iterrows():
        job_id = row['id']
        job_description = row['combined_text']

        job_skills = extract_skills(job_description, skill_set)
        matched_skills = [skill for skill in job_skills if skill in user_skills]
        missing_skills = [skill for skill in job_skills if skill not in user_skills]

        result[job_id] = [matched_skills, missing_skills]

    return result


def main_with_skills(job_title=None, job_location=None, num_outputs=None, filepath='sample_resume.pdf'):
    # Extract jobs and preprocess
    if ',' in job_location:
        job_area, job_country = job_location.strip().split(',')
    else:
        job_area, job_country = job_location, job_location

    jobs = scraper(jobtitle=job_title, pref_area=job_area, no_of_output=num_outputs)
    new_jobs = preprocess_scraped_data(jobs)

    # Extract and preprocess resume text
    resume_text = ResumeReader(filepath)
    user_skills = extract_skills(preprocess_resume(resume_text))

    # Get recommended jobs
    recommended_jobs = predict(new_jobs, resume_text)

    # Match skills
    skill_matches = match_skills(recommended_jobs, user_skills)

    # Display results
    for job_id, (has_skills, lacks_skills) in skill_matches.items():
        print(f"Job ID: {job_id}")
        print(f"  - Skills you have for this job: {has_skills}")
        print(f"  - Skills you lack for this job: {lacks_skills}")

    return skill_matches
```
